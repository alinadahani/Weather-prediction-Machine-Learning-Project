### Encoding

```python
#Encoding the column Precip type

df['Precip Type']=df['Precip Type'].replace('rain',1)
df['Precip Type']=df['Precip Type'].replace('snow',0)
df
```

```python
#Merging all the other classes  into 4 classes depending on their similarity such such "Partly cloudy " to "cloudy"
df['Summary']=df['Summary'].replace('Partly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Mostly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Overcast','Cloudy')
df['Summary']=df['Summary'].replace('Foggy','Windy')
df['Summary']=df['Summary'].replace('Breezy and Overcast','Windy')
df['Summary']=df['Summary'].replace('Breezy and Mostly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Breezy and Partly Cloudy','Windy')
df['Summary']=df['Summary'].replace('Dry and Partly Cloudy','Clear')
df['Summary']=df['Summary'].replace('Windy and Partly Cloudy','Windy')
df['Summary']=df['Summary'].replace('Light Rain','Rain')
df['Summary']=df['Summary'].replace('Breezy','Windy')
df['Summary']=df['Summary'].replace('Windy and Overcast','Windy')
df['Summary']=df['Summary'].replace('Humid and Mostly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Drizzle','Rain')
df['Summary']=df['Summary'].replace('Breezy and Foggy','Windy')
df['Summary']=df['Summary'].replace('Windy and Mostly Cloudy','Windy')
df['Summary']=df['Summary'].replace('Dry','Clear')
df['Summary']=df['Summary'].replace('Humid and Partly Cloudy','Clear')
df['Summary']=df['Summary'].replace('Dry and Mostly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Mostly Cloudy','Cloudy')
df['Summary']=df['Summary'].replace('Humid and Overcast','Cloudy')
df['Summary']=df['Summary'].replace('Windy and Foggy','Windy')
df['Summary']=df['Summary'].replace('Windy and Dry','Windy')
df['Summary']=df['Summary'].replace('Dangerously Windy and Partly Cloudy','Windy')
df['Summary']=df['Summary'].replace('Breezy and Dry','Windy')

df
```

### Handling Outliers

```python
Handle outliers in numerical features
numerical_features = ['Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)',
                      'Visibility (km)', 'Precip Type', 'Pressure (millibars)']
threshold = 3

for feature in numerical_features:
    mean = np.mean(df[feature])
    std_dev = np.std(df[feature])
    df = df[(df[feature] - mean) <= threshold * std_dev]
df
```
